```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN
```

```python
sentence=["I","Love","Deep","Learning"]
print(sentence)
```

```python
word_embeddings={
    "I":1.0,
    "Love":2.0,
    "Deep":3.0,
    "Learning":4.0
}
print(word_embeddings)
```

```python
x=np.array([
    [
    [word_embeddings["I"]],
    [word_embeddings["Love"]],
    [word_embeddings["Deep"]],
    [word_embeddings["Learning"]]
    ]
])
print(x)
```

```python
y=np.array([word_embeddings["Learning"]])
print(y)
print("Input shape:",x.shape)
print("Output shape:",y.shape)
```

```python
#simpleRnn model
model=Sequential()
model.add(SimpleRNN
          (units=8,activation='tanh',input_shape=(3,1)))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam',loss='mse')
```

```python
#train model
model.fit(x,y,epochs=300,verbose=1)
```

```python
#prediction
test_sequence=np.array([
 [
    [1.0],
    [2.0],
    [3.0]
]
])
prediction = model.predict(test_sequence)
print("prediction:",prediction[0][0])
```

```python
#convert prediction to closest word
predicted_number=prediction[0][0]
closest_word=min(word_embeddings,key=lambda word:abs(word_embeddings[word]-predicted_number))
print("Predicted word:",closest_word)
```
